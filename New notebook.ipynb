```
from IPython.display import display, Image
def show_comics(number):
    import requests
    response = requests.get(f"https://xkcd.com/{number}/info.0.json")
    resp = response.json()
    print (f"Заголовок: {resp['safe_title']}")
    print (f"Дата публикации: {resp['day']} {resp['month']} {resp['year']}")
    display(Image(f"{resp['img']}"))
    print (f"Подпись: \n{resp['alt']}")



#show_comics(614)
#show_comics(1589)
#show_comics(2509)
show_comics(1987)
```
